Objective: ask the user to enter the basic Pay ,
<pre>
if basicPay <=5000 , then calculate HRA as 16% of basicPay , and DA as 14% of basicPay

if basicPay >5000, then Caclulate HRA as 12% of basicPay and DA as 10% of basicPay

find gross Pay i.e. sum of basicPay + HRA + DA

```python

basicPay = int (input('enter basic pay 💴 '))

if basicPay > 5000 :

  HRA= .12 * basicPay
  DA = .1  * basicPay

else:

  HRA = .16 * basicPay
  DA  = .14 * basicPay

# let us find the grossPay

grossPay = basicPay + HRA + DA

# let us use string formatting using F-String to print the salary details

salarySummary =f'''
-------------------------Salary Slip -----------------
BasicPay : {basicPay}
HRA     :  {HRA}
DA      :  {DA}
-------------------
Gross :    {grossPay}
'''

print( salarySummary )
```

```python
text='abcdefghi'
```

i can see the indexes , can u also see the indexes?

```python
# Variable[ StartIndex :  EndIndex   ] # End Index is never included

text[ 1 : 4 ]        # i.e. from 1 index till 3 index
```

```python
text
```

```python
# variable [   :   EndIndex  ]  # when skipping Start index, default is 0 index

text[ : 5 ]    # i.e. from 0 index till 4 index
```

```python
text
```

```python
# variable [ StartIndex :    ]   # when skipping end index , default is length-1 index

text[ 5 : ]         # i.e from 5th index till length -1 index i.e. 9-1  8th index
```
